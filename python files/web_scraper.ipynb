```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm


df  = pd.read_csv('scotus_texts.csv')
df.head()
```

```python
a = df.url
data_arr = []
for index,url in tqdm(enumerate(a)) :
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        p_text = soup.find_all('p')
        temp_text = ''
        for i in range(len(p_text)):
            if len(p_text[i].get_text()) > 50 :
                temp_text = temp_text + p_text[i].get_text()
        data_arr.append([index,temp_text])
    except :
        continue
        
        
    
    
    
    
```

```python
 
import csv       
fields = ['Index', 'Corpus']
rows = data_arr    # ENTER THE DATA CORPUS HERE 
filename = "data.csv"
with open(filename, 'w') as csvfile:  
    csvwriter = csv.writer(csvfile)  
    csvwriter.writerow(fields)  
    csvwriter.writerows(rows) 

```
